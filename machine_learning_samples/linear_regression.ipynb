```python
# samples from w3schools and other files

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression
```

```python
n = 20
X = np.arange(n)
y = 6*X + 3*(X**2) + 2*(X**3) -100
plt.scatter(X,y)
```

```python
m = LinearRegression()
m.fit(X[:,np.newaxis],y)
y_pred = m.predict(X[:,np.newaxis])
plt.scatter(X,y)
plt.plot(X,y_pred,color='red')
```

```python
from sklearn.preprocessing import PolynomialFeatures

#degree - którego stopnia krzywa dopasowania 1 - liniowa,2 - kwadratowa,3

polyModel = PolynomialFeatures(degree=3,include_bias=True)
X_poly = polyModel.fit_transform(X[:,np.newaxis])

m = LinearRegression()

m.fit(X_poly,y)


y_pred = m.predict(X_poly)
plt.scatter(X,y)
plt.plot(X,y_pred,color='red')
```

```python
m.coef_
#współczynniki - x^0 x^1 x^2 x^3 ...
```

```python
n = 20
X = np.arange(n)
y = 3*X-4 + 10*(np.random.randn(X.size))
plt.scatter(X,y)
```

```python
#tworzenie krzywej najlepszego dopasowania
polyModel = PolynomialFeatures(degree=5)
X_poly = polyModel.fit_transform(X[:,np.newaxis])
m = LinearRegression()
m.fit(X_poly,y)
y_pred = m.predict(X_poly)
plt.scatter(X,y)
plt.plot(X,y_pred,color='red')
```

```python
# współczynnik korelacji
corr = np.corrcoef(y,y_pred)
print(corr)
```

```python
np.random.seed(2)

#liczba 100 reprezentuje liczebność

x = np.random.normal(3, 1, 100)
y = np.random.normal(150, 40, 100) / x

plt.scatter(x, y)
plt.show()
```

```python
#80% na trenowanie ; 20% na testowanie próbek
train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

mymodel = np.poly1d(np.polyfit(train_x, train_y, 4))

#linia od lini 0 do lini 5 ; 10 łamanych
myline = np.linspace(0, 5, 10)

plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline), color='red')
plt.show()
```

```python
#współczynnik regresji

from sklearn.metrics import r2_score

r2 = r2_score(train_y, mymodel(train_x))

print(r2)

#sprawdź czy testowe wartosci też są ok

r3 = r2_score(test_y, mymodel(test_x))

print(r3)
#wartości podobne więc jest model poprawny
```
