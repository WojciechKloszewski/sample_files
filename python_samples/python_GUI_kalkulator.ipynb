```python
# based on Geeksforgeeks file
# calculator - Python , GUI 

from tkinter import *
 
# global declaration of the expression variable/deklaracja zmiennej wynikowej(początkowo wartość pusta)
expression = ""
 
    ## Update expression/zmień wyrażenie (do obliczeń)
# text entry box / teks we wstępnym oknie - formuła, wynik
def press(num):
    # global expression variable/ określ zmienną globalną
    global expression
 
    # concatenation of string / konkatenacja stringa - liczy wynik z funkcji łańcuchowej 
    expression = expression + str(num)
 
    # update the expression (using set method) / zmiana wartości używając metody set
    equation.set(expression)
 
 
# Function to evaluate the final expression / funkcja ukazująca wyniki

def equalpress():
    # Try and except - handling errors like divide by zero to detect it /
    # użycie try i except aby wykryć błędy typu dzielenie przez zero

 
    # code inside the try block where generate the error/
    # kod wewnątrz "pudełka" gdzie jest generowany błąd
    try:
 
        global expression
 
        # eval function evaluate the expression, convert the result to string / 
        # fukcja wyniku total, konwersja wyrażenia na string

        total = str(eval(expression))
        equation.set(total)
 
        # initialize the expression variable by empty string /
        # inicjalizacja wyrażenia przez pusty string
        expression = ""
 
    # if generate error - handle by the except block/
    # jeśli wygeneruje błąd to złap go do bloku wyjątków
    except: 
        equation.set(" error ")
        expression = ""
 
 # Clear the contents function of text entry box /
 # wyczyść zawartość funkcji tekstowej w entry box - długi prostokąt
def clear():
    global expression
    expression = ""
    equation.set("")
 
 
# Main code / kod główny
if __name__ == "__main__":
    
    # GUI window 
    gui = Tk()
 
    # background colour of GUI window / kolor tła okna
    gui.configure(background="light green")
 
    # title of GUI window / określ tytuł okna
    gui.title("Kalkulator")
 
    # configuration of dimensions GUI window / wymiary okna kalkulatora
    gui.geometry("500x260")
 
    # StringVar() is the variable class , create an instance of this class /
    # StringVar() jako zmienna klasowa , utworzenie instancji tej klasy
    equation = StringVar()
 
    # create text entry box to show the expression/
    # stworzenie okna wyniku kalkulatora by pokazać wyniki (string)
    expression_field = Entry(gui, textvariable=equation)
 
    # grid method used to place the widget at respective positions (table) /
    # metoda siatki by umieścić widgety na odpowiednich pozycjach (tabelarycznie)

    expression_field.grid(columnspan=4, ipadx=170, ipady=8)
 
    # create buttons and grid position, color, dimensions /
    # utwórz przyciski - pozycja na siatce, kolor, wymiary
    
    # event function - press, equalpress, clear (trigger)/
    # funkcja zdarzenia - press, equalpress, clear , podobne do trigger
    
    button1 = Button(gui, text=' 1 ', fg='black', bg='green',
                    command=lambda: press(1), height=2, width=15)
    button1.grid(row=2, column=0)
 
    button2 = Button(gui, text=' 2 ', fg='black', bg='green',
                    command=lambda: press(2), height=2, width=15)
    button2.grid(row=2, column=1)
 
    button3 = Button(gui, text=' 3 ', fg='black', bg='green',
                    command=lambda: press(3), height=2, width=15)
    button3.grid(row=2, column=2)
 
    button4 = Button(gui, text=' 4 ', fg='black', bg='green',
                    command=lambda: press(4), height=2, width=15)
    button4.grid(row=3, column=0)
 
    button5 = Button(gui, text=' 5 ', fg='black', bg='green',
                    command=lambda: press(5), height=2, width=15)
    button5.grid(row=3, column=1)
 
    button6 = Button(gui, text=' 6 ', fg='black', bg='green',
                    command=lambda: press(6), height=2, width=15)
    button6.grid(row=3, column=2)
 
    button7 = Button(gui, text=' 7 ', fg='black', bg='green',
                    command=lambda: press(7), height=2, width=15)
    button7.grid(row=4, column=0)
 
    button8 = Button(gui, text=' 8 ', fg='black', bg='green',
                    command=lambda: press(8), height=2, width=15)
    button8.grid(row=4, column=1)
 
    button9 = Button(gui, text=' 9 ', fg='black', bg='green',
                    command=lambda: press(9), height=2, width=15)
    button9.grid(row=4, column=2)
 
    button0 = Button(gui, text=' 0 ', fg='black', bg='green',
                    command=lambda: press(0), height=2, width=15)
    button0.grid(row=5, column=0)
 
    plus = Button(gui, text=' + ', fg='black', bg='green',
                command=lambda: press("+"), height=2, width=15)
    plus.grid(row=2, column=3)
 
    minus = Button(gui, text=' - ', fg='black', bg='green',
                command=lambda: press("-"), height=2, width=15)
    minus.grid(row=3, column=3)
 
    multiply = Button(gui, text=' * ', fg='black', bg='green',
                    command=lambda: press("*"), height=2, width=15)
    multiply.grid(row=4, column=3)
 
    divide = Button(gui, text=' / ', fg='black', bg='green',
                    command=lambda: press("/"), height=2, width=15)
    divide.grid(row=5, column=3)
 
    equal = Button(gui, text=' = ', fg='black', bg='green',
                command=equalpress, height=2, width=15)
    equal.grid(row=5, column=2)
 
    clear = Button(gui, text='Clear', fg='black', bg='green',
                command=clear, height=2, width=15)
    clear.grid(row=5, column='1')
 
    Decimal= Button(gui, text='.', fg='black', bg='green',
                    command=lambda: press('.'), height=2, width=15)
    Decimal.grid(row=6, column=0)
    # start the GUI
    gui.mainloop()
```
