```python
%matplotlib notebook

import pandas as pd
import matplotlib.pyplot as plt
from itertools import count
import random
from matplotlib.animation import FuncAnimation

x_vals = []
y_vals = []

y1=[1,4,5,2,1,7]

index = count()

def animate(i):
    
    x_vals.append(next(index))
    y_vals.append(y1[i]) 

    
    plt.cla()
    plt.plot(x_vals, y_vals)
    
ani = FuncAnimation(plt.gcf() , animate , interval=1000 )   
    
plt.plot(x_vals, y_vals)

plt.tight_layout()
plt.show()
```

```python
import random
import matplotlib
from matplotlib import animation
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib notebook


l1 = [random.randint(-5,2)+(i**1.83)/(random.randint(13,14)) for i in range(0,80,2)]
l2 = [random.randint(-10,-5)+(i**1.6)/(random.randint(10,12)) for i in range(0,80,2)]
l3 = [random.randint(-20,-10)-(i**1.3)/(random.randint(11,12)) for i in range(0,80,2)]
root_list = [random.randint(-80,-75)+(i**1.5)/(random.randint(10,11)) for i in range(0,80,2)]

l1 = root_list + l1
l2 = root_list + l2
l3 = root_list + l3

plt.style.use("seaborn")

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (10,5))
axes.set_ylim(-300, 500)
axes.set_xlim(1980, 2060)
plt.xlabel("testowe kategorie")
plt.ylabel("testowe dane") 

x1,y1,y2,y3 = [], [], [], []
def animate(i):
    
    x1.append(1980+i)
    y1.append((l1[i]))
    y2.append((l2[i]))
    y3.append((l3[i]))
    
    axes.plot(x1,y1, color="red", linewidth=0.5)    
    axes.plot(x1,y3, color="blue", linewidth=0.5)
    axes.plot(x1,y2, color="gray", linewidth=1)
    
anim = FuncAnimation(fig, animate, interval=50)
```
