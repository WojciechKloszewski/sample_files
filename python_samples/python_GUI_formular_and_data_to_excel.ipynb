```python
# based on Geeksforgeeks file

from openpyxl import *
from tkinter import *
  
# opening the existing excel file or create file and change path/
#otwórz istniejący plik excel lub stwórz plik i zmień ścieżkę na właściwą
wb = load_workbook('C:\\Users\\Wojciech Kłoszewski\\Desktop\\test.xlsx')
 
# create the sheet object/ stworzyć obiket arkusz (excel)
sheet = wb.active
 
def excel():
     
    # resize the width of columns in excel spreadsheet /
    # zmień szerokość kolumn w arkuszu excel
    sheet.column_dimensions['A'].width = 35
    sheet.column_dimensions['B'].width = 12
    sheet.column_dimensions['C'].width = 12
    sheet.column_dimensions['D'].width = 22
    sheet.column_dimensions['E'].width = 22
    sheet.column_dimensions['F'].width = 42
    sheet.column_dimensions['G'].width = 52
 
    # write given data to an excel spreadsheet/
    # napisz podane informacje jako nagłówki kolumn w arkuszu excel
    sheet.cell(row=1, column=1).value = "Name"
    sheet.cell(row=1, column=2).value = "Course"
    sheet.cell(row=1, column=3).value = "Semester"
    sheet.cell(row=1, column=4).value = "Form Number"
    sheet.cell(row=1, column=5).value = "Contact Number"
    sheet.cell(row=1, column=6).value = "Email id"
    sheet.cell(row=1, column=7).value = "Address"
 
 
# Function to set focus (cursor) and set focus on the course_field box/
# ustaw kursor w arkuszu excel na odpowiednią komórkę
def focus1(event):
    course_field.focus_set()

def focus2(event):
    sem_field.focus_set()

def focus3(event):
    form_no_field.focus_set()

def focus4(event):
    contact_no_field.focus_set()

def focus5(event):
    email_id_field.focus_set()

def focus6(event):
    address_field.focus_set()

# Clear the contents of text entry boxes - empty cells /
# wyczyść obecną zawartość komórek excel

def clear():
     
    # clear the content of text entry box/ wyczyść zawartość pustych komórek
    name_field.delete(0, END)
    course_field.delete(0, END)
    sem_field.delete(0, END)
    form_no_field.delete(0, END)
    contact_no_field.delete(0, END)
    email_id_field.delete(0, END)
    address_field.delete(0, END)

# Function to take data from GUI window and write to an excel file/
# funkcja do pobrania danych z okna GUI i wpisanie do pliku excel
def insert():
     
    # empty field NULL - if user not fill any entry, print "empty input"/
    # puste pole NULL - jeśli użytkownik nie wypełni pola wpisz "empty input"
    if (name_field.get() == "" and
        course_field.get() == "" and
        sem_field.get() == "" and
        form_no_field.get() == "" and
        contact_no_field.get() == "" and
        email_id_field.get() == "" and
        address_field.get() == ""):
             
        print("empty input")
 
    else:
 
        # assigning the max row and max column value up to which data is written in an excel sheet to the variable/
        # określ które wiersze, kolumny zawierają wartości w pliku excel ; by przejść do kolejnego wiersza po submicie
        current_row = sheet.max_row
        current_column = sheet.max_column
 
        # get method returns current text as string and write into excel - next excel row/
        # metoda która zwraca obecnie zaznaczony tekst jako string a nastepnie zapisuje do excela w kolejnym exelowym wierszu

        sheet.cell(row=current_row + 1, column=1).value = name_field.get()
        sheet.cell(row=current_row + 1, column=2).value = course_field.get()
        sheet.cell(row=current_row + 1, column=3).value = sem_field.get()
        sheet.cell(row=current_row + 1, column=4).value = form_no_field.get()
        sheet.cell(row=current_row + 1, column=5).value = contact_no_field.get()
        sheet.cell(row=current_row + 1, column=6).value = email_id_field.get()
        sheet.cell(row=current_row + 1, column=7).value = address_field.get()
 
        # save the file / zapisz plik (określ ścieżkę)
        wb.save('C:\\Users\\Wojciech Kłoszewski\\Desktop\\test.xlsx')
 
        # set focus on the name_field box/ ustaw zaznaczenie danej komórki w pliku excel
        name_field.focus_set()
 
        # clear() function in text box in GUI/ funkcja czyszcząca dane z pól formularza w GUI
        clear()
 
 
# Main code / kod główny
if __name__ == "__main__":
     
    # GUI window / okno programu
    root = Tk()
 
    # background colour of GUI window / kolor tła
    root.configure(background='light green')
 
    # title of GUI window / tytuł okna
    root.title("registration form")
 
    # configuration of GUI window / konfiguracja okna programu
    root.geometry("500x300")
 
    excel()
 
    # Form label
    heading = Label(root, text="Form", bg="light green")
 
    # Name label
    name = Label(root, text="Name", bg="light green")
 
    # Course label
    course = Label(root, text="Course", bg="light green")
 
    # Semester label
    sem = Label(root, text="Semester", bg="light green")
 
    #  Form No. label
    form_no = Label(root, text="Form No.", bg="light green")
 
    # Contact No. label
    contact_no = Label(root, text="Contact No.", bg="light green")
 
    # Email id label
    email_id = Label(root, text="Email id", bg="light green")
 
    # address label
    address = Label(root, text="Address", bg="light green")
 
    # grid method for placing the widgets /
    # metoda siatki do umiejscowienia widżetów

    heading.grid(row=0, column=1)
    name.grid(row=1, column=0)
    course.grid(row=2, column=0)
    sem.grid(row=3, column=0)
    form_no.grid(row=4, column=0)
    contact_no.grid(row=5, column=0)
    email_id.grid(row=6, column=0)
    address.grid(row=7, column=0)
 
    # text entry box for typing the information in GUI /
    # umiejsowienie pół tekstowych w GUI do wpisania danych
    name_field = Entry(root)
    course_field = Entry(root)
    sem_field = Entry(root)
    form_no_field = Entry(root)
    contact_no_field = Entry(root)
    email_id_field = Entry(root)
    address_field = Entry(root)
 
    # bind method of widget is used for
    # the binding the function with the events/ event function - wiązanie przycisków z funkcjami
 
    name_field.bind("<Return>", focus1)

    course_field.bind("<Return>", focus2)

    sem_field.bind("<Return>", focus3)

    form_no_field.bind("<Return>", focus4)

    contact_no_field.bind("<Return>", focus5)

    email_id_field.bind("<Return>", focus6)
 
    # grid method fo for placing widgets/
    # metoda siatki do umiejscowienia widżetów
    name_field.grid(row=1, column=1, ipadx="100")
    course_field.grid(row=2, column=1, ipadx="100")
    sem_field.grid(row=3, column=1, ipadx="100")
    form_no_field.grid(row=4, column=1, ipadx="100")
    contact_no_field.grid(row=5, column=1, ipadx="100")
    email_id_field.grid(row=6, column=1, ipadx="100")
    address_field.grid(row=7, column=1, ipadx="100")
 
    # call excel function/ wywołaj funkcję exel
    excel()
 
    # Submit Button and place into the root window / przycisk potwierdzenia i umiejscowienie go w oknie
    submit = Button(root, text="Submit", fg="Black",
                            bg="Red", command=insert)
    submit.grid(row=8, column=1)
 
    # start the GUI
    root.mainloop()
```
