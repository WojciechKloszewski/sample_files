```python
frame = pd.read_csv("dane/Mcdonalds.csv")
frame.head()
```

```python
#wstawianie wartoÅ›ci
frame.loc[1,'Calories']=np.NaN
frame.loc[2,'TotalFat']=np.NaN
frame.loc[4,'Category']=np.NaN
frame.head()
```

```python
frame.dropna().head()
```

```python
#usuniÄ™cie tylko tych pozycji ktÃ³re sÄ… wszystkie puste
frame.dropna(how='all').head()
```

```python
#usuniÄ™cie tylko tych pozycji ktÃ³re majÄ… puste wartoÅ›ci w podanych kolumnach
frame.dropna(subset=['TotalFat']).head()
```

```python
#usuwamy kolumny z zerami axis=1 ; wiersze axis=0 ; inplace=True - zapisz
frame.dropna(axis=1).head()
```

```python
#wypeÅ‚nij puste dane fillna wartoÅ›ciÄ… 0
frame.fillna(value=0).head()
```

```python
dictionary = {'Category':'UNKNOWN','Calories':0 , 'TotalFat':0}
frame.fillna(value=dictionary).head()
```

```python
#inny sposÃ³b dla danej kolumny
frame['Category'].fillna(value='UNKNOWN' , inplace=True)
```

```python
#wypeÅ‚nienie wartoÅ›ciÄ… Å›redniÄ…
avg=frame['TotalFat'].mean()
frame['TotalFat'].fillna(avg, inplace=True)
```
