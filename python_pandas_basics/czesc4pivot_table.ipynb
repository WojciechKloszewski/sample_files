```python
#tabela przestawna
sales = pd.read_csv('dane/WA_Sales_Products_2012-14.csv')
sales.head()
```

```python
#tabela przestawna
sales.pivot_table(index='Retailer country', columns=['Year','Quarter'],values="Revenue").head()
```

```python
#tabela przestawna
sales.pivot_table(index='Retailer country', columns=['Year'],values="Revenue").head()
```

```python
#tabela przestawna
sales.pivot_table(index='Retailer country', columns=['Year'],values="Gross margin" , aggfunc=["min","max"]).head()
```

```python
#jakie są nagłówki tabeli
pivTab = sales.pivot_table(index='Retailer country', columns=['Year'],values="Gross margin" , aggfunc=["min","max"]).head()
pivTab.columns
```

```python
#zamienić rzędy w kolumnie by rok był na wierzchu
pivTab = pivTab.swaplevel(axis='columns').sort_index(axis='columns')
pivTab.head()
```

```python
#melt
#nowy pivot table
pt = sales.pivot_table(index="Retailer country", columns="Year", values="Revenue",aggfunc="sum")
pt.head()
```

```python
pt.reset_index(inplace=True)
pt.fillna(0,inplace=True)
pt.head()
```

```python
pt.melt(id_vars="Retailer country", value_name="Revenuesum", var_name="YearOfTransaction").head()
```

```python
#nowy pivot table
pt = sales.pivot_table(index="Retailer country", columns="Year", values="Revenue",aggfunc="sum")
```

```python
pt.head()
```

```python
#odpivotowanie
upvt = pt.unstack().to_frame().swaplevel().head()
upvt.columns=["Revenue_sum"]
upvt.head()
```
