```python
cat1 = pd.read_csv("dane/categories_del_1.csv", usecols=['CategoryID','CategoryName'])
cat2 = pd.read_csv("dane/categories_del_2.csv", usecols=['CategoryID','CategoryName'])
```

```python
cat1
```

```python
cat2
```

```python
#czym różnią się dwie tabele
#innej join
cat1.merge(cat2)
```

```python
#pokazuje wszystko co jest po lewej stronie czyli cat1, sufiksy dodają końcówki do nazw dodatkowych kolumn, kolumna lewa , prawa
cat1.merge(cat2, how="left", on="CategoryID", suffixes=['_lewa','_prawa'])
```

```python
#wszystko "lewo i prawo" ; indicator - pokazuje gdzie jest jakaś dana, w którym zbiorze
cat_merged = cat1.merge(cat2, how="outer", on="CategoryID", suffixes=['_lewa','_prawa'], indicator=True)
cat_merged
```

```python
#filtorwanie wartości, które były tylko po lewej stronie
filtr = cat_merged["_merge"] == "left_only"
cat_merged[filtr]
```

```python
#znalezienie tych produktów, które się nie sprzedały
products = pd.read_csv("dane/products.csv", usecols=['ProductID','ProductName'])
products.head()
```

```python
orderedProducts = pd.read_csv("dane/order-details_del.csv", usecols=['OrderID','ProductID','UnitPrice','Quantity'])
orderedProducts.head()
```

```python
merged = products.merge(right=orderedProducts, on="ProductID", how="outer",indicator=True)
merged.head()
```

```python
#określ rodzaje złączeń
merged['_merge'].value_counts()
```

```python
filtr = merged['_merge']=='left_only'
```

```python
#pokazanie niesprzedanych produktów
merged[filtr]
```

```python
merged = products.merge(right=orderedProducts, on="ProductID", how="left",indicator=True)
```

```python
merged['_merge'].value_counts()
```

```python
#inna wersja tego samego - zamiana zbiorów oraz right
merged = orderedProducts.merge(right=products, on="ProductID", how="right",indicator=True)
merged['_merge'].value_counts()
```

```python
#pokazanie niesprzedanych produktów
filtr = merged['_merge']=='right_only'
merged[filtr]
```

```python
products = pd.read_csv("dane/products.csv", usecols=['ProductID','ProductName','CategoryID','UnitPrice'])
categories = pd.read_csv("dane/categories1.csv")
```

```python
products.head()
```

```python
categories.head()
```

```python
products.merge(right=categories, on='CategoryID').head()
```

```python
#zmiana nazwy kolumny - aby użyć join, merge: np. zmień nazwę kolmuny
categories.rename({'CategoryID':'ID'}, axis='columns', inplace=True)
categories.head()
```

```python
products.merge(right=categories, left_on='CategoryID', right_on='ID')
```

```python
products.rename({'ProductName':'Name'}, axis='columns', inplace=True)
categories.rename({'CategoryName':'Name'}, axis='columns', inplace=True)
```

```python
#na szczęście problem z tymi samymi nazwami kolumn rozwiązywane są automatycznie
products.merge(right=categories, left_on='CategoryID', right_on='ID', suffixes=['_product','_category'], sort=True)
```

```python
categories.set_index('ID',inplace=True)
```

```python
#by działał merge po indexie, dodaj polecenie na index ; jest też left_index
products.merge(right=categories, left_on='CategoryID', right_index=True , suffixes=['_product','_category'], sort=True)
```
